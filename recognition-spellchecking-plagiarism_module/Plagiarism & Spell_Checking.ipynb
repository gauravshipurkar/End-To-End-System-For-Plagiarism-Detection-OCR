```python
class BhuvanaSpellCorrector:

    def __init__(self, word_list):

        self.word_list = word_list
        self.new_line = " "
        self.result = []
        self.spell()
        self.result.append(self.new_line)

    def correct(self, inputs, model, tokenizer):
        input_ids = tokenizer.encode(inputs, return_tensors='pt')
        sample_output = model.generate(
            input_ids,
            do_sample=True,
            max_length=50,
            top_p=0.99,
            num_return_sequences=1
        )
        res = tokenizer.decode(sample_output[0], skip_special_tokens=True)
        return res

    def spell(self):
        tokenizer = AutoTokenizer.from_pretrained("Bhuvana/t5-base-spellchecker")
        model = AutoModelForSeq2SeqLM.from_pretrained("Bhuvana/t5-base-spellchecker")
        for word in self.word_list:
            word = ''.join(word)
        print(word)

        self.word_list = ' '.join(self.word_list)
        self.new_line = self.correct(self.word_list, model, tokenizer)
        print(self.new_line)

        return self.new_line
```

```python
def correct(inputs, model, tokenizer):
  input_ids = tokenizer.encode(inputs, return_tensors='pt')
  sample_output = model.generate(
      input_ids,
      do_sample=True,
      max_length=50,
      top_p=0.99,
      num_return_sequences=1
  )
  res = tokenizer.decode(sample_output[0], skip_special_tokens=True)
  return res
```

# Plagiarism 


```python
import os
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sample_files = [doc for doc in os.listdir('/content/lol/') if doc.endswith('.txt')]
print(sample_files)
sample_contents = [open('/content/lol/'+File).read() for File in sample_files]
vectorize = lambda Text: TfidfVectorizer().fit_transform(Text).toarray()
similarity = lambda doc1, doc2: cosine_similarity([doc1, doc2])

vectors = vectorize(sample_contents)
```

```python
s_vectors = list(zip(sample_files, vectors))
```

```python
def check_plagiarism():
  results = set()
  global s_vectors
  for sample_a, text_vector_a in s_vectors:
    new_vectors = s_vectors.copy()
    current_index = new_vectors.index((sample_a, text_vector_a))
    del(new_vectors[current_index])
    for sample_b, text_vector_b in new_vectors:
      sim_score = similarity(text_vector_a, text_vector_b)[0][1]
      sample_pair = sorted((sample_a, sample_b))
      score = sample_pair[0], sample_pair[1], sim_score
      print(score)
      results.add(score)
  return results 

for data in check_plagiarism():
  print(data)
```

```python
from sentence_transformers import SentenceTransformer
from nltk.tokenize import sent_tokenize
from sklearn.metrics.pairwise import cosine_similarity 
from pprint import pprint
import torch
```

```python
import nltk
nltk.download('punkt')
```

```python
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
with open('/content/lol/file1.txt') as handle:
  file1 = handle.read()

with open('/content/lol/file2.txt') as handle:
  file2 = handle.read()
  
```

```python
sentences_file1 = file1.split('.')
sentences_file2 = file2.split('.')
```

```python
sentence2=[]
sentence1=[]
```

```python
for sentence in sentences_file2:
  sentence = sentence.replace("\n", "")
  sentence = sentence.replace("\t","")
  sentence = sentence.replace("•","")
  if sentence == '':
    continue
  sentence2.append(sentence)

for sentence in sentences_file1:
  sentence = sentence.replace("\n", "")
  sentence = sentence.replace("\t","")
  sentence = sentence.replace("•","")
  if sentence == '':
    continue
  sentence1.append(sentence)
```

```python
print(sentence2)
```

```python
print(sentence1)
```

```python
sentence_embeddings1 = model.encode(sentence1)
sentence_embeddings2 = model.encode(sentence2)
```

```python
for sentence, embedding in zip(sentence1, sentence_embeddings1):
  print('Sentence',sentence)
  print('Embedding',embedding)
```

```python
results = []
```

```python
for i in range(0,len(sentence1)):
  for j in range(0,len(sentence2)):
    results.append([sentence1[i],sentence2[j],cosine_similarity(sentence_embeddings1[i].reshape(1,-1),sentence_embeddings2[j].reshape(1,-1))[0][0]])
```

```python
for res in results:
  if (res[2]*100)>=50:
    print(res)
```
